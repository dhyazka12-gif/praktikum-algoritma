Dictionary

```python
# a. buat dictionary keranjang
keranjang = {
    "apel": 5000,
    "pisang": 3000,
    "jeruk": 7000,
    "anggur": 15000
}

# b. gunakan perulangan untuk menghitung semua total harga barang
total = 0
for harga in keranjang.values():
    total += harga

# c. tampilkan setiap harga dan barangnya
for barang, harga in keranjang.items():
    print(f"barang {barang}: harga = {harga}")

# d. tampilkan total harga
print(f"Total harga: {total}")
```
