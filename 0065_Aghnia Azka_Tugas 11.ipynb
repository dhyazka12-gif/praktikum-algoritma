```python
def tambah_buku():
    print("\n=== Tambah Buku Baru ===")
    key = input("Masukkan KODE/KEY buku: ")

    judul = input("Masukkan Judul Buku: ")
    penulis = input("Masukkan Nama Penulis: ")
    tahun = input("Masukkan Tahun Terbit: ")
    tanggal_pinjam = input("Masukkan Tanggal Terakhir Dipinjam (dd-mm-yyyy): ")
    status = input("Masukkan Status (Tersedia/Tidak): ")

  # Menyimpan data buku
    perpustakaan[key] = {
        "Judul Buku": judul,
        "Penulis": penulis,
        "Tahun Terbit": tahun,
        "Tanggal Terakhir Pinjam": tanggal_pinjam,
        "Status": status
    }
    print("\nBuku berhasil ditambahkan!\n")

def tampilkan_buku():
    """Fungsi untuk menampilkan semua data buku"""
    print("\n=== Daftar Buku di Perpustakaan ===")
    if not perpustakaan:
        print("Belum ada data buku yang tersimpan.")
    else:
        for key, data in perpustakaan.items():
            print(f"\nKEY: {key}")
            print(f"Judul Buku           : {data['Judul Buku']}")
            print(f"Penulis              : {data['Penulis']}")
            print(f"Tahun Terbit         : {data['Tahun Terbit']}")
            print(f"Tanggal Terakhir Pinjam : {data['Tanggal Terakhir Pinjam']}")
            print(f"Status               : {data['Status']}")
    print("-----------------------------------")

perpustakaan = {}
while True:
    print("\n=== MENU PERPUSTAKAAN ===")
    print("1. Tambah Buku")
    print("2. Tampilkan Semua Buku")
    print("3. Keluar")

    pilihan = input("Pilih menu (1/2/3): ")

    if pilihan == "1":
        tambah_buku()
    elif pilihan == "2":
        tampilkan_buku()
    elif pilihan == "3":
        print("Terima kasih! Program selesai.")
        break
    else:
        print("Pilihan tidak valid. Coba lagi.")
```
