```python
# latihan

print("1. Luas persegi panjang")
print("2. Volume balok")
print("3. Luas lingkaran")
print("4. Volume bola")

pilihan = input("Pilih operasi (1/2/3/4): ")

if pilihan == "1":
  panjang = int(input("Masukan panjang: "))
  lebar = int(input("Masukkan lebar: "))
  Luas_persegi_panjang = panjang * lebar
  print(f"Luas persegi panjang = {Luas_persegi_panjang}")

elif pilihan == "2":
  panjang = int(input("Masukan panjang: "))
  lebar = int(input("Masukkan lebar: "))
  tinggi = int(input("Masukan tinggi: "))
  Volume_balok = panjang * lebar * tinggi
  print(f"Volume balok = {Volume_balok}")

elif pilihan == "3":
  jari_jari = float(input("Masukan jari-jari: "))
  # Luas_lingkaran = 3,14 * jari-jari
  Luas_lingkaran = 3.14 * jari_jari * jari_jari
  print(f"Luas lingkaran = {Luas_lingkaran}")

elif pilihan == "4":
  jari_jari = float(input("Masukan jari-jari: "))
  # Volume_bola = 4/3 * 3,14 * jari-jari
  Volume_bola = 4/3 * 3.14 * jari_jari * jari_jari
  print(f"Volume bola = {Volume_bola}")

else:
  print("Pilihan tidak valid")
```
