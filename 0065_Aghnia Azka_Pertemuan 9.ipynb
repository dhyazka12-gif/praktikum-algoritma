```python
## soal nomor 1
buah = ["apel", "mangga", "jeruk", "anggur", "pisang"]
print(buah)

# tambahkan semangka pada list akhir
buah.append("semangka")
print(f"data sebelum ditambah : {buah}\n")

# sisipkan durian di antara jeruk dan anggur
buah.insert(3, "durian")
print(f"data setelah ditambah : {buah}\n")

# hapus mangga dari list
buah.remove("mangga")
print(f"data setelah dihapus : {buah}\n")

# ubah pisang menjadi nanas
buah[4] = "nanas"
print(f"setelah diubah : {buah}\n")

# tampilkan tiga buah pertama
print(buah[:3])
```

```python
## soal nomor 2
angka = [45, 12, 78, 23, 56, 89, 34]

# urutkan data secara ascending
angka.sort()
print(f"setelah diurutkan : {angka}\n")

# urutkan secara descending
angka.reverse()
print(f"setelah urutan dibalik : {angka}")
```
